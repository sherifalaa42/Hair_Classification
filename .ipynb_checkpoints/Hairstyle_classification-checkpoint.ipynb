```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.keras import models, layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
train_gen = ImageDataGenerator(rescale= 1./255)

train_ds = train_gen.flow_from_directory(
    '/content/data/train',
    target_size= (200,200),
    class_mode = 'binary',
    batch_size= 20,
    shuffle= True,
    seed = SEED
)

test_gen = ImageDataGenerator(rescale= 1./255)

test_ds = test_gen.flow_from_directory(
    '/content/data/test',
    target_size= (200,200),
    batch_size=20,
    shuffle= True,
    class_mode = 'binary'
)
```

```python
model = models.Sequential(
    [
        keras.Input(shape= (200,200,3)),
        layers.Conv2D(32, (3,3), activation= 'relu'),
        layers.MaxPool2D(pool_size= (2,2)),

        layers.Flatten(),

        layers.Dense(64, activation= 'relu'),
        layers.Dense(1, activation= 'sigmoid')
    ]
)
model.summary()
```

```python
model.layers
```

```python
my_callbacks = [
    keras.callbacks.ModelCheckpoint(
        'best_model_ {epoch:02d}_{val_accuracy:.3f}.h5',
        save_best_only= True,
        monitor= 'val_accuracy',
        mode= 'max'
    )
]
```

```python
learining_rate = 0.002
momentum = 0.8

optimizer = keras.optimizers.SGD(learning_rate= learining_rate, momentum= momentum)

loss = keras.losses.BinaryCrossentropy()

model.compile(optimizer= optimizer, loss= loss, metrics= ['accuracy'])
```

```python
history = model.fit(train_ds, epochs=10, validation_data= test_ds)
```

```python
history.history
```

```python
plt.plot(history.history['accuracy'], label='ACC')
plt.plot(history.history['val_accuracy'], label='VAL')
plt.ylim(.65,.8)
plt.legend()
plt.show()
```

```python
np.median(np.array(history.history['accuracy']))
```

```python
np.std(np.array(history.history['loss']))
```

```python
train_gen_aug = ImageDataGenerator(
    rescale= 1./255,
    rotation_range=50,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)
train_ds_aug = train_gen_aug.flow_from_directory(
    '/content/data/train',
    target_size= (200,200),
    class_mode = 'binary',
    batch_size= 20,
    shuffle= True,
    seed = SEED
)
```

```python
train_ds.class_indices
```

```python
history_aug = model.fit(train_ds_aug,
                        epochs=10,
                        validation_data= test_ds,
                        callbacks = my_callbacks)
```

```python
plt.plot(history_aug.history['accuracy'], label='ACC')
plt.plot(history_aug.history['val_accuracy'], label='VAL')
plt.xticks(np.arange(10))
plt.legend()
plt.show()
```
